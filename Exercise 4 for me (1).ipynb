```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
sb.set() # set the default Seaborn style for graphics
# Import mean_squared_error from sklearn
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
train = pd.read_csv('train.csv')
```

```python
SalePrice =  pd.DataFrame(train["SalePrice"])
GrLivArea =  pd.DataFrame(train["GrLivArea"])
SalePrice.head()
```

```python
SalePrice_train, SalePrice_test,GrLivArea_train, GrLivArea_test = train_test_split(SalePrice,GrLivArea, test_size = 0.20)
```

```python
linreg = LinearRegression() 
linreg.fit(GrLivArea_train,SalePrice_train )     
print('Intercept\t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
```

```python
regline_x = GrLivArea_train
regline_y = linreg.intercept_ + linreg.coef_ *GrLivArea_train

# Plot the Linear Regression line
f = plt.figure(figsize=(16, 8))
plt.scatter(GrLivArea_train, SalePrice_train)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()
GrLivArea_train_pred = linreg.predict(GrLivArea_train)
print("Explained Variance (R^2) \t:", linreg.score(GrLivArea_train, SalePrice_train))

# Mean Squared Error (MSE)
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))


mse = mean_sq_err(GrLivArea_train, GrLivArea_train_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

```python
# Predict Total values corresponding to HP Test
GrLivArea_test_pred = linreg.predict(GrLivArea_test)

# Plot the Predictions
f = plt.figure(figsize=(16, 8))
plt.scatter(GrLivArea_test, SalePrice_test, color = "green")
plt.scatter(GrLivArea_test, GrLivArea_test_pred, color = "red")
plt.show()
# Explained Variance (R^2)
print("Explained Variance (R^2) \t:", linreg.score(GrLivArea_test, SalePrice_test))

# Mean Squared Error (MSE)
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))

mse = mean_sq_err(SalePrice_test, GrLivArea_test_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

By observing the Explained Variance from all three models, the first model which is SalePrice and GrLivArea seems to be the best one since its R^2 value is the highest with 0.5 .
```python
trains = train[["GrLivArea", "SalePrice"]]
Q1 = trains.quantile(0.25)
Q3 = trains.quantile(0.75)
IQR = Q3 - Q1
# remove outliers from dataset
df = trains[~((trains < (Q1 - 1.5 * IQR)) |(trains > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df.shape)
# any is if condition
Sale_train, Sale_test , GrLivArea1_train, GrLivArea1_test = train_test_split(df[["SalePrice"]],df[["GrLivArea"]], test_size=0.2)
linreg = LinearRegression()
linreg.fit(GrLivArea1_train, Sale_train)

print('Intercept \t: b = ', linreg.intercept_)
print('Coefficients \t: a = ', linreg.coef_)

# Formula for the Regression line
regline_x = GrLivArea1_train
regline_y = linreg.intercept_ + linreg.coef_ * GrLivArea1_train

# Plot the Linear Regression line
f = plt.figure(figsize=(16, 8))
plt.scatter(GrLivArea1_train, Sale_train)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()

GrLivArea1_train_pred = linreg.predict(GrLivArea1_train)
# Check the Goodness of Fit (on Train Data)
print("Explained Variance (R^2) \t:", linreg.score(GrLivArea1_train, Sale_train))
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))

mse = mean_sq_err(Sale_train, GrLivArea1_train_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

```python
GrLivArea2_test_pred = linreg.predict(GrLivArea1_test)

# Plot the Predictions
f = plt.figure(figsize=(16, 8))
plt.scatter(GrLivArea1_test, Sale_test, color = "green")
plt.plot(GrLivArea1_test, GrLivArea2_test_pred, color = "red")
plt.show()

# Explained Variance (R^2)
print("Explained Variance (R^2) \t:", linreg.score(GrLivArea1_test, Sale_test))
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))
mse = mean_sq_err(Sale_test, GrLivArea2_test_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

By comparing the R^2 values from Prob 1  with Prob 3(with removed outliers) , Prob 1 R^2 value is better for both Train and Test than the Prob 3 R^2 value. The values for MSE and RMSE in the  Prob 3 dataset has reduced for both test and train, this means that the second model is better because lower RMSE values indicate a better fit. For prediction, RMSE is a good measure and it is the most important criterion for fit.
